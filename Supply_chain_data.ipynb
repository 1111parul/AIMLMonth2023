```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
data = [
    112.6, 32.9, 13.36, 106.88, 129.14, 74.25, 171.22, 6.18, 77.25, 27.9,
    15.45, 43.17, 16.45, 33.9, 24.72, 60.8, 19.54, 43.26, 9.27, 46.35, 65.8
]
labels = [f'Weight {i+1}' for i in range(len(data))]
x = np.arange(len(data))
slope, intercept, r_value, p_value, std_err = linregress(x, data)
linear_fit = slope * x + intercept
plt.figure(figsize=(12, 6))
plt.scatter(x, data, label='Data Point ', color='b')
plt.plot(x, linear_fit, label='Linear Fit', linestyle='-', color='r')
plt.xlabel('Weights')
plt.ylabel('Values')
plt.title('Scatter Plot with Linear Regression')
plt.xticks(x, labels, rotation=90)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [
    101.7, 66.8, 84.7, 78.34, 46.35, 89.7, 83.4, 50.68, 15.4, 48.2,
    31.26, 16.4, 14.4, 7.7, 16.4, 6.16, 42.2, 38.04, 14.4, 16.4, 32.8
]


labels = [f'Weight {i+1}' for i in range(len(data))]


plt.figure(figsize=(12, 6))
plt.scatter(labels, data, color='b', label='Data Point ')


x_values = np.arange(len(data))
coefficients = np.polyfit(x_values, data, 1)
y_fit = np.polyval(coefficients, x_values)
plt.plot(labels, y_fit, color='r', linestyle='-', label='Linear Regression')

plt.xlabel('Weights')
plt.ylabel('Values')
plt.title('Scatter Plot with Linear Regression')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()


plt.grid(True)
plt.show()
```
